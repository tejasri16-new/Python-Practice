```python
for num in range(50,101):
    count=0
    for i in range(1,num+1):
        if num % i ==0:
            count += 1
    if count ==2 :
        print(num)
```

```python
def prime(n):
    if n<2:
        print("Not a Prime number")
        return
    for i in range(2,n):
        if n % i ==0:
            print("Not a Prime number")
            return
    print("Prime number")
num = int(input("Enter a number:"))
prime(num)
```
